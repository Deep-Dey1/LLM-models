```python
from datasets import load_dataset

# Load dataset (you can replace this with your own dataset if you have one)
dataset = load_dataset("openwebtext")
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load GPT-2 model and tokenizer
model_name = "gpt2"  # You can also use "gpt2-medium" or "gpt2-large" for larger models
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Add padding token since GPT-2 does not have one by default
tokenizer.pad_token = tokenizer.eos_token
```

```python
def tokenize_function(examples):
    return tokenizer(examples["text"], return_tensors="pt", truncation=True, padding="max_length")

# Tokenize the dataset
tokenized_datasets = dataset.map(tokenize_function, batched=True)
tokenized_datasets.set_format("torch")  # Set format for PyTorch
```
